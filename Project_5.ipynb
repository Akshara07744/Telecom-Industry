What are the value counts of the target variable


```python
x = df1["Churn"].value_counts()
plt.title("-------- Churn Count--------")
plt.pie(x,labels=x.index,autopct="%1.1f%%")
plt.show()
```

```python
figure = plt.figure(figsize=(11,7))
plt.subplot(2,2,1)   #nrows,ncols,index
sns.countplot(x="InternetService",hue="Churn",data=df1,edgecolor="black")

plt.subplot(2,2,3)
sns.countplot(x="MultipleLines",hue="Churn",data=df1,edgecolor="black")

plt.subplot(2,2,2)
plt.title("=====Distibution by Partner=====")
countval = df1["Partner"].value_counts()
plt.pie(countval,labels=countval.index,autopct="%1.1f%%",colors=["r","g"])

plt.subplot(2,2,4)
plt.title("=====Distibution by Gender=====")
countgender = df1["gender"].value_counts()
plt.pie(countgender,labels=countgender.index,autopct="%1.1f%%",colors=["m","b"])

plt.show()
```
